# **Project Summary -**

Customer satisfaction is one of the most important factors that determines the success of any e-commerce platform. In a highly competitive online retail environment, companies like Flipkart must ensure that customers receive quick and effective support when they face issues related to orders, deliveries, payments, or product returns. A positive customer support experience can significantly improve customer loyalty, while poor support service may lead to dissatisfaction and customer churn. Therefore, analyzing customer satisfaction data and understanding the factors that influence it is crucial for improving service quality and overall business performance.

This project focuses on analyzing customer support interaction data to understand the factors that affect Customer Satisfaction Scores (CSAT). The dataset contains information about various customer support interactions, including details such as issue categories, support channels used by customers, response times from support agents, product categories, and other operational attributes. Each record represents a customer interaction along with the CSAT score provided by the customer after the issue was resolved. These scores help measure how satisfied customers are with the support service they received.

The first stage of the project involves performing Exploratory Data Analysis (EDA) to gain insights into the dataset and identify patterns that influence customer satisfaction. During this stage, data cleaning and preprocessing are carried out to handle missing values, incorrect data types, and inconsistencies in the dataset. Various visualization techniques and statistical methods are used to explore relationships between different variables and the CSAT score. For example, the analysis investigates how response time affects customer satisfaction, whether certain issue categories lead to lower satisfaction scores, and how different support channels such as chat, email, or phone influence the overall customer experience.

Through EDA, meaningful insights are generated about operational factors that may contribute to higher or lower customer satisfaction. For instance, longer response times may result in lower CSAT scores, while faster issue resolution may improve customer experience. Similarly, certain issue types or product categories might require better support processes. Identifying these patterns helps businesses understand where improvements are needed in the customer support system.

In the next stage of the project, machine learning techniques are applied to build a predictive model that can estimate customer satisfaction scores based on the available features in the dataset. By training classification models on historical support interaction data, the system can learn patterns that influence customer satisfaction. Once trained, the model can predict whether a customer interaction is likely to result in a high or low satisfaction score.

This predictive capability can help businesses proactively identify potentially dissatisfied customers and take corrective actions before the situation escalates. For example, support teams can prioritize cases that are more likely to lead to dissatisfaction or improve processes that frequently result in low CSAT scores. Ultimately, this project demonstrates how data analysis and machine learning can be used to enhance customer support performance, improve customer experience, and support data-driven decision making in an e-commerce environment.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Create a copy of dataset
df_clean = df.copy()

# Convert date columns to datetime
df_clean['Issue_reported at'] = pd.to_datetime(df_clean['Issue_reported at'], errors='coerce')
df_clean['issue_responded'] = pd.to_datetime(df_clean['issue_responded'], errors='coerce')

# Create response time feature
df_clean['response_time_minutes'] = (
    df_clean['issue_responded'] - df_clean['Issue_reported at']
).dt.total_seconds() / 60

# Drop columns with extremely high missing values
df_clean = df_clean.drop(columns=['connected_handling_time'])

# Check dataset after wrangling
df_clean.head()
````

**Data Wrangling Steps**

- Created a copy of the dataset to perform cleaning operations.

- Converted timestamp columns Issue_reported_at and issue_responded to datetime format.

- Generated a new feature response_time_minutes by calculating the difference between issue response time and issue reported time.

- Removed the column connected_handling_time because it contained extremely high missing values.

- Prepared the dataset for further exploratory data analysis.

### What all manipulations have you done and insights you found?

**Insights from Data Wrangling**

- Timestamp columns can be used to derive meaningful features like response time, which may influence customer satisfaction.

- Some columns contain excessive missing values and may not contribute meaningfully to analysis.

- Creating derived features from existing data can help reveal patterns affecting CSAT scores.
